```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv('data_eur_2019_fr.csv',sep=',',dtype ={'Cn':'object','IT':'object'})
#je force le type de ces 2 colonnes car certaines valeurs pourraient être interprétées comme numériques
display(df.head())
display(df.info())
display(df.describe())
```

```python
#le dataframe ayant 36 colonnes j'autorise l'affichage de 36 colonnes
pd.set_option('display.max_columns', 36)
```

```python
#import du dictionnaire de données
dico=pd.read_csv("dico_donnees.csv")
dico
```

```python
#renommage des colonnes pour leur donner des noms plus usuels
col_df=pd.DataFrame({'Name':df.columns})
colonnes=col_df.merge(dico, on ='Name')
col_liste=colonnes['Nom'].values
df.set_axis(col_liste, axis='columns', inplace=True)
df.head()
```

```python
#L'Id étant la clef primaire je le passe en index
df=df.set_index('ID')
df.head()
```

```python
#Je vérifie les doublons absolus
print("Nombre de duplicate",df.duplicated().sum())                  
#suppression des doublons
df=df.drop_duplicates(keep = 'first')    
print("Nombre de duplicate", df.duplicated().sum())                   
```

```python
df.info()
```

```python
#Je supprime les champs Autonomie_elec et Reduc_emission_techno_WLPT qui ne sont jamais remplis
df.drop(['Autonomie_elec','Reduc_emission_techno_WLPT','Nb_enregistrement'],axis=1, inplace=True)
#Je supprime le champs CO2_NEDC car ce n'est plus la norme en vigueur
df.drop(['CO2_NEDC'],axis=1, inplace=True)
```

```python
#je regarde si certaines variables sont systématiquement nulles pour les voitures électriques,
#auquel cas je remplacerai les NaN par -1 pour les numériques car c'est une valeur non utilisée (cf describe) et 'elec' pour les textuelles
df[df['Type_carburant']=='electric'].info()
```

```python
df.loc[df['Type_carburant']=='electric','Cylindree']=-1
df.loc[df['Type_carburant']=='electric','Reduc_emission_techno']=-1
df.loc[df['Type_carburant']=='electric','Facteur_deviation']=-1
df.loc[df['Type_carburant']=='electric','Facteur_verif']=-1
df.loc[df['Type_carburant']=='electric','Techno_innovante']='elec'
```

```python
df[df['Type_carburant']=='electric'].describe()
```

```python
#Comme il y a assez peu de modèles électriques dans la base, je vais gérer les manquants de facon spécifique pour eux :
#Pour les manquants de Test_WLPT_Mass : je prends la valeur de la Masse roulante
df.loc[(df['Type_carburant']=='electric') & (df['Test_WLPT_Mass'].isnull()),'Test_WLPT_Mass']=df['Masse_roulante']

#Pour les manquants de la Consommation moyenne et de la puissance : je prends la médiane car l'écart-type est trop grand pour prendre la moyenne
df.loc[(df['Type_carburant']=='electric') & (df['Consommation_elec'].isnull()),'Consommation_elec']=df[df['Type_carburant']=='electric']['Consommation_elec'].median()
df.loc[(df['Type_carburant']=='electric') & (df['Puissance'].isnull()),'Puissance']=df[df['Type_carburant']=='electric']['Puissance'].median()

#Pour les manquants VFN et groupe constructeur je mets "elec"
df.loc[(df['Type_carburant']=='electric') & (df['VFN'].isnull()),'VFN']='elec'
df.loc[(df['Type_carburant']=='electric') & (df['Groupe_constructeur'].isnull()),'Groupe_constructeur']='elec'
display(df[df['Type_carburant']=='electric'].describe())
display(df[df['Type_carburant']=='electric'].info())
```

```python
#Pour estimer le coût en CO2 d'une voiture électrique je multiplis sa consommation en Wh/km par le coût moyen du Wh en CO2 dans le pays
#J'importe donc la table de correspondance : pays_CO2/Wh
#Lorsqu'on utilisera le modèle, pour retrouver l'émission en CO2 d'une voiture électrique il faudra faire une conversion :
#le modèle nous donnera son émission en CO2 au coût français, il faudra faire une règle de 3 pour retrouver son émission dans le pays concerné
#on pourra en profiter pour afficher sa consommation électrique : 
# CO2 prédit / CO2/WhFR = conso électrique prédite
#conso électrique prédite * CO2/Wh pays = émission CO2 prédite adaptée au pays.
Pays_CO2=pd.read_csv('CO2_elec.csv',decimal=",",dtype ={'Pays':'object','CO2/kWh':'float64','CO2/Wh':'float64'})

display(Pays_CO2)
df=df.merge(Pays_CO2,on='Pays', how='left')
df.loc[df['Type_carburant']=='electric','CO2_WLPT']=df['Consommation_elec']*df['CO2/Wh']
df[df['Type_carburant']=='electric'].describe()

#je ne fais pas l'exercice pour les hybrides car je considère qu'elles se rechargent en roulant donc le coût en CO2 du rechargement est déjà compris dans l'émission de CO2 WLPT.
```

```python
#Vérfication du poids des manquants restants
manquants=df.isnull().sum()/len(df.index)*100 
manquants.sort_values(ascending = False)
```

```python
#D'après le dictionnaire de données, ce que nous cherchons à prédire c'est la variable CO2_WLPT, en effet c'est celle qui est utilisée en Europe.
#Je supprime donc toutes les lignes pour lesquelles cette variable n'est pas renseignée. 
df.dropna(subset=['CO2_WLPT'], inplace=True)


#Pour les autres variables : je supprime les lignes avec des vides si le % de Nan est <5% (ayant plus de 200 000 données je peux me le permettre)
Nan5=manquants[(manquants>0) & (manquants<5)].index
df.dropna(subset=Nan5, inplace=True)


#Je supprime les champs statuts et année car je n'ai pris que les données finales de 2019
#Je supprime le champs Pays car je n'ai pris que la France et j'ai déjà fait l'enrichissement pour les voitures électriques
df.drop(['Statut','Annee','Pays'],axis=1, inplace=True)

#Je supprime les variables CO2/Wh et CO2/kWh qui ne servaient que pour le calcul de consommation des voitures électriques
df.drop(['CO2/kWh','CO2/Wh'],axis=1, inplace=True)

#Je vérifie le résultat
df.info()
```

```python
#je regarde s'il y a des doublons sur les champs sans manquant et je les supprime
df.iloc[:,0:24].duplicated().sum()
index_doublons=df[df.iloc[:,0:22].duplicated()==True].index
print("Nombre de doublons sur les colonnes sans manquant ",len(index_doublons))
df.drop(index_doublons , inplace=True)
df.info()
```
